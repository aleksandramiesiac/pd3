# Trzy zbiorki danych z ilustracjami

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from spectral import spectral_clustering
```

```python
# zaladowanie plikow
lista_baz = ["zbiorek1","zbiorek2","zbiorek3"]

zbiorek1 =  np.array(pd.read_csv("moje/"+lista_baz[0]+".data").iloc[:,[1,2]])
zbiorek2 =  np.array(pd.read_csv("moje/"+lista_baz[1]+".data").iloc[:,[1,2]])
zbiorek3 =  np.array(pd.read_csv("moje/"+lista_baz[2]+".data").iloc[:,[1,2]])
```

```python
plt.scatter(zbiorek1.transpose()[0],zbiorek1.transpose()[1])
plt.show()
```

```python
wynik1 = spectral_clustering(zbiorek1, k=2, M=10)
plt.scatter(zbiorek1.transpose()[0],zbiorek1.transpose()[1], c=wynik1)
plt.show()
```

```python
plt.scatter(zbiorek2.transpose()[0],zbiorek2.transpose()[1])
plt.show()
```

```python
wynik2 = spectral_clustering(zbiorek2, k=2, M=10)
plt.scatter(zbiorek2.transpose()[0],zbiorek2.transpose()[1],c =wynik2)
plt.show()
```

```python
plt.scatter(zbiorek3.transpose()[0],zbiorek3.transpose()[1])
plt.show()
```

```python
wynik3 = spectral_clustering(zbiorek3, k=2, M=5)

plt.scatter(zbiorek3.transpose()[0],zbiorek3.transpose()[1], c=wynik3)
plt.show()
```
